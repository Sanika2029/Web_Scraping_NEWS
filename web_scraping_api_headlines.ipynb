```python
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'data'
# Count the number of headlines in each section
section_counts = datanew['sectionname'].value_counts()

# Create a bar graph
plt.figure(figsize=(10, 6))
section_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Headlines by Section')
plt.xlabel('Section Name')
plt.ylabel('Number of Headlines')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from textblob import TextBlob

# Function to calculate sentiment
def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity  # Returns a value between -1 (negative) and 1 (positive)

# Apply the sentiment function to the webtitle column
datanew['sentiment'] = datanew['webtitle'].apply(get_sentiment)

# Display the first few rows with sentiment
print(datanew[['webtitle', 'sentiment']].head())

# Optional: Visualize the sentiment distribution
plt.figure(figsize=(10, 6))
plt.hist(datanew['sentiment'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Sentiment Distribution of Headlines')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
